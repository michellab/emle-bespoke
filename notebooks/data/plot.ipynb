```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for better aesthetics
sns.set(style="whitegrid", palette="muted")

# Dictionary of data file paths
data = {
    #"bespoke1_iter_1_reference": "bespoke2_iter_1_reference.csv",
    "bespoke1_iter_2_reference": "bespoke2_iter_2_reference.csv",
    #"bespoke1_iter_3_reference": "bespoke2_iter_3_reference.csv",
    #"bespoke2_iter_1_reference": "bespoke2_iter_1_reference.csv",
    #"bespoke2_iter_2_reference": "bespoke2_iter_2_reference.csv",
    #"bespoke2_iter_3_reference": "bespoke2_iter_3_reference.csv",
    #"bespoke1_iter_1_species": "bespoke1_iter_1_species.csv",
    #"bespoke1_iter_2_species": "bespoke1_iter_2_species.csv",
    #"bespoke1_iter_3_species": "bespoke1_iter_3_species.csv",
    #"bespoke2_iter_1_species": "bespoke2_iter_1_species.csv",
    #"bespoke2_iter_2_species": "bespoke2_iter_2_species.csv",
    #"bespoke2_iter_3_species": "bespoke2_iter_3_species.csv",
    #"patched_species1_iter_1_species": "patched_species1_iter_1_species.csv",
    #"patched_species1_iter_2_species": "patched_species1_iter_2_species.csv",
    #"patched_species1_iter_3_species": "patched_species1_iter_3_species.csv",
    #"patched_species2_iter_1_species": "patched_species2_iter_1_species.csv",
    #"patched_species2_iter_2_species": "patched_species2_iter_2_species.csv",
    #"patched_species2_iter_3_species": "patched_species2_iter_3_species.csv",
    #"patched_reference2_iter_1_reference": "patched_reference2_iter_1_reference.csv",
    #"patched_reference2_iter_2_reference": "patched_reference2_iter_2_reference.csv",
    #"patched_reference1_iter_3_reference": "patched_reference1_iter_3_reference.csv",
    #"patched_reference2_iter_3_reference": "patched_reference2_iter_3_reference.csv",
}

# Create an empty DataFrame to collect the data
all_data = []

# Loop through each file and extract relevant columns
for label, filepath in data.items():
    # Load the CSV file into a pandas DataFrame
    df = pd.read_csv(filepath)
    
    # Ensure 'mol' and 'rmse' columns exist
    if 'mol' in df.columns and 'mse' in df.columns:
        # Create a temporary DataFrame with the label
        df_temp = df[['mol', 'mse']].copy()
        df_temp['label'] = label  # Add a column to identify the dataset
        all_data.append(df_temp)

# Combine all individual DataFrames into one
final_data = pd.concat(all_data, ignore_index=True)

# Create a beautiful bar plot using Seaborn
plt.figure(figsize=(14, 8))  # Increase the size for readability
sns.barplot(x='mol', y='rmse', hue='label', data=final_data, palette='coolwarm', width=0.8)

# Set plot labels and title
plt.xlabel("Mol", fontsize=16)
plt.ylabel("RMSE", fontsize=16)
plt.title("Mol vs RMSE for Various Datasets", fontsize=18)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=12)

# Improve legend placement
plt.legend(title="Datasets", title_fontsize=14, loc='upper left', bbox_to_anchor=(1, 1), fontsize=12)

# Add a grid for easier readability
plt.grid(True, axis='y', linestyle='--', linewidth=0.7)

# Adjust layout to prevent clipping of labels
plt.tight_layout()

# Display the plot
plt.show()
```
